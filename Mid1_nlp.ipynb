```python
import nltk
import spacy
import gensim
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam
from gensim.models import Word2Vec, FastText
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import time

# NLTK resources download
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# NLP tools initializeaztion
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

# Sample text
text_data = [
    "Natural Language Processing enables computers to understand human language.",
    "Word embeddings represent words in dense vectors.",
    "Neural networks can be used for training embeddings like Word2Vec, GloVe, and FastText.",
    "Text preprocessing includes tokenization, stopword removal, and lemmatization.",
    "Deep learning methods improve NLP tasks with contextualized word embeddings."
]

# Text preprocessing
def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text.lower())
    # Lemmatization & Stopword removal
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha() and word not in stop_words]  
    return tokens

# Applying preprocessing
preprocessed_texts = [preprocess_text(text) for text in text_data]

# Training of Word2Vec and FastText embeddings
cbow_model = Word2Vec(sentences=preprocessed_texts, vector_size=50, window=5, min_count=1, sg=0)
fasttext_model = FastText(sentences=preprocessed_texts, vector_size=50, window=5, min_count=1, sg=1)

# Visualization of word embeddings
def visualize_embeddings(model, method='tsne', title="Word Embeddings Visualization"):
    # Selection of top words
    words = list(model.wv.index_to_key)[:min(30, len(model.wv.index_to_key))]  
    vectors = np.array([model.wv[word] for word in words])
    
    if method == 'tsne':
        # Ensure perplexity is valid
        perplexity = min(5, len(words) - 1)  
        reduced_vectors = TSNE(n_components=2, perplexity=perplexity).fit_transform(vectors)
    else:
        reduced_vectors = PCA(n_components=2).fit_transform(vectors)
    
    plt.figure(figsize=(8, 6))
    plt.scatter(reduced_vectors[:, 0], reduced_vectors[:, 1], color='blue')
    for i, word in enumerate(words):
        plt.annotate(word, (reduced_vectors[i, 0], reduced_vectors[i, 1]), fontsize=9)
    plt.title(title)
    plt.show()

# Fig 1: t-SNE visualization of Word2Vec embeddings
visualize_embeddings(cbow_model, method='tsne', title="Figure 1: t-SNE Visualization of Word2Vec Embeddings")

# Fig 2: PCA visualization of FastText embeddings
visualize_embeddings(fasttext_model, method='pca', title="Figure 2: PCA Visualization of FastText Embeddings")

# Loading of IMDb dataset (built-in TensorFlow dataset)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=20000)
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=200)
x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=200)

# Creation of RNN model
def create_rnn_model():
    model = Sequential([
        Embedding(input_dim=20000, output_dim=128, input_length=200),
        SimpleRNN(64, return_sequences=False, dropout=0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Training of RNN model
rnn_model = create_rnn_model()
history_rnn = rnn_model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))

## Fig 3: RNN training accuracy vs. loss curve
plt.figure(figsize=(8, 5))
plt.plot(history_rnn.history['accuracy'], label='Train Accuracy')
plt.plot(history_rnn.history['val_accuracy'], label='Validation Accuracy')
plt.title("Figure 3: RNN Training Accuracy vs. Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# Creation of LSTM model
def create_lstm_model():
    model = Sequential([
        Embedding(input_dim=20000, output_dim=128, input_length=200),
        LSTM(64, dropout=0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Creation of GRU model
def create_gru_model():
    model = Sequential([
        Embedding(input_dim=20000, output_dim=128, input_length=200),
        GRU(64, dropout=0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Training of LSTM and GRU models
lstm_model = create_lstm_model()
gru_model = create_gru_model()

start_time = time.time()
history_lstm = lstm_model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))
lstm_time = time.time() - start_time

start_time = time.time()
history_gru = gru_model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))
gru_time = time.time() - start_time

# Fig 5: LSTM vs. GRU accuracy comparison
plt.figure(figsize=(8, 5))
plt.plot(history_lstm.history['accuracy'], label='LSTM Train Accuracy')
plt.plot(history_gru.history['accuracy'], label='GRU Train Accuracy')
plt.title("Figure 5: LSTM vs. GRU Accuracy Comparison")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# Fig 9: Bidirectional LSTM vs. Standard LSTM
bidirectional_lstm_model = Sequential([
    Embedding(input_dim=20000, output_dim=128, input_length=200),
    Bidirectional(LSTM(64, dropout=0.3, return_sequences=True)),
    Bidirectional(LSTM(64, dropout=0.3)),
    Dense(1, activation='sigmoid')
])

bidirectional_lstm_model.compile(optimizer=Adam(learning_rate=0.0005), loss='binary_crossentropy', metrics=['accuracy'])
history_bilstm = bidirectional_lstm_model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))

plt.figure(figsize=(8, 5))
plt.plot(history_bilstm.history['accuracy'], label='Bidirectional LSTM Accuracy')
plt.plot(history_lstm.history['accuracy'], label='Standard LSTM Accuracy')
plt.title("Figure 9: Comparison of Bidirectional LSTM and Standard LSTM Model Performance")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()


```
